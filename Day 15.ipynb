```python
# Benrnoulli  Distribution - p^x(1-p)^1-x
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
p = 0.7 # Probability of success 
x = [0,1] # Possible outcomes
# Probability Mass Function(pmf)
# Probability Density Function(PDF)
pmf = bernoulli.pmf(p,x)
plt.scatter(x,pmf,color = "red")
plt.title("Benoulli Bistribution")
plt.xlabel("Benoulli's pmf Value")
plt.ylabel("The POssible Outcome")
```

```python
# Binomial Distribution - n(c^x)(p^x)(q^(n-x))

from scipy.stats import binom
import numpy as np 
n = 10
p = 0.5
x = np.arange(0,n+1) # We can use For instead for how many time we wnat to check.  It checks each values from it
pmf = binom.pmf(x,n,p) # pmf = mass function it gives the plotin points
plt.scatter (x,pmf,color = "Red")
plt.xlabel("The pmf value is")
plt.ylabel("The Outcome is")
plt.title("Binomial Distribution")
plt.show()
print(pmf)
```

```python
# Poission Distribution - Its Done By Interval Time
# To calaculate  the average possible outcomes with in the interval of time 
# (e^-lambda * lambda^k)/n!

from scipy.stats import poisson
import numpy as np
lambda_value = 5
k = np.arange(0,10)
pmf = poisson.pmf(lambda_value,k)
plt.bar(k,pmf,color = "red")
plt.show()
```

```python
# Continues Distribution - Infinite Measurable
# Include - Integration, Differenciation, etc..............
'''
1.uniform distribution - 1/b-a
'''

from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
a=2
b=5
x=np.linspace(a,b,200) #its for giving the spacing 
pdf=uniform.pdf(x,loc=a,scale=b-a)
print(pdf)
plt.bar(x,pdf,color="yellow")
plt.show()
```

```python
# exponential - (Same as Poisson distribution) average rate     (lambda*e^ -lambda x)


from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

lambda_value=2
x=np.linspace(0,10,50)
pdf=expon.pdf(x,scale=1/lambda_value) #exponential mean value - 1/lambda
print(pdf)
plt.bar(x,pdf,color="pink")
plt.show()
```

```python
'''
mean - E(x)- summation (xi * pi)
variance - E(X^2)-E(X)^2
            

zip- to iterate multiple list present in a folder
'''
def mean_variance(x,x_prob):
    E=sum(v*p for v,p in zip(x,x_prob))   #E(x)
    E2=sum((v**2)*p for v,p in zip(x,x_prob)) #E(x)^2(summation (xi)^2 *(pi))
    # for v,p in zip(x,x_prob)
            # S1=SUM(V)*P
    # for v,p in zip(x,x_prob):
            # s2=sum(v**2)*p

    var_value=E2-E**2  #E(x^2)-E(x)^2
    return E,var_value

x=[0,1,2,3]
x_prob=[0.3,0.5,0.2]
E,V=mean_variance(x,x_prob)
print(f"The mean value is {E} and the variance is {V}")
```

```python
# Central limit theorem
# dies roll=30 times,trails=1000times, find the mean and plot the graph

import numpy as np
import matplotlib.pyplot as plt
import random

# parameters
no_trails=1000
no_dice=30
sample_mean=[]

# iteration

for i in range(no_trails):
    rolls=np.random.randint(1,7,no_dice)
    sample_mean.append(np.mean(rolls))

# plots

plt.hist(sample_mean,bins=30,color="green",density=True)
plt.title("Central Limit Theorem With dice Rolls")
plt.xlabel("sample mean of 30 roll")
plt.ylabel("density")
plt.grid(True)
plt.axvline(x=3.5,color="red",linestyle="--",label="Expected Mean(3.5)")
plt.legend()
plt.show()
```
