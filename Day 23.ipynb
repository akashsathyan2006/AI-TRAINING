```python
# Assignment 5  - 7/8/2025

# 1. Organizational Hierarchy (Tree View)
# Scenario:
# A company’s reporting structure:
#           CEO
#          /   \
#     Manager1 Manager2
#      /             \
#   Employee1     Employee2
# Why Binary Tree?
# •	Each person can report to one manager (parent), and may have up to 2 direct subordinates.
# Use-case:
# •	Traversals (level order, in-order) can be used to generate org charts or send emails level-wise.



class Node():
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None

root = Node("CEO")
root.left = Node("Manager1")
root.right = Node("Manager2")
root.left.left = Node("Employee1")
root.right.right = Node("Employee2")


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print("    " * level, node.val)
        print_tree(node.left, level + 1)

print_tree(root)
```

```python
# 2. Tournament Bracket System
# Scenario:
# Any knockout-style tournament:
#        Final
#       /     \
#   Semi1     Semi2
#    /  \     /  \
#  Q1  Q2   Q3  Q4
# Why Binary Tree?
# •	Each match has two players (left/right).
# •	The winner moves up as the parent node (next match).
# Use-case:
# •	Winner determination, path to final, etc.



class Node():
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

q1 = Node("Q1")
q2 = Node("Q2")
q3 = Node("Q3")
q4 = Node("Q4")

semi1 = Node("Semi1")
semi1.left = q1
semi1.right = q2

semi2 = Node("Semi2")
semi2.left = q3
semi2.right = q4

final = Node("Final")
final.left = semi1
final.right = semi2


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)


print_tree(final)
```
